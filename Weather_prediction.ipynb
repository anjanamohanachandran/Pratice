```python
from sklearn.linear_model import Ridge,LinearRegression
from sklearn.metrics import mean_absolute_error,accuracy_score
```

```python
model_1=Ridge(alpha=.1)
```

```python
predictors=["precip","temp_max","temp_min"]
train=core_weather.loc[:"2020-12-31"]
test=core_weather.loc["2020-12-31":,:]
```

```python
model_1.fit(train[predictors],train["target"])
predictions=model_1.predict(test[predictors])
model_1.score(test[predictors],test["target"])
```

```python
error_model_1=mean_absolute_error(test["target"],predictions)
error_model_1
```

```python
combined=pd.concat([test["target"],pd.Series(predictions,index=test.index,name="predictions")],axis=1)
combined.columns=["actual","predictions"]
combined.plot()
```

```python
model_1.coef_
```

```python
def weather_predictions(predictors,core_weather,model_1):
    train=core_weather.loc[:"2020-12-31"]
    test=core_weather.loc["2020-12-31":]
    model_1.fit(train[predictors],train["target"])
    predictions=model_1.predict(test[predictors])
    error=mean_absolute_error(test["target"],predictions)
    combined=pd.concat([test["target"],pd.Series(predictions,index=test.index,name="predictions")],axis=1)
    combined.columns=["actual","predictions"]
    return error,combined
```

```python
core_weather["month_max"]=core_weather["temp_max"].rolling(30).mean()
```

```python
core_weather=core_weather[30:]
```

```python
core_weather
```

```python
core_weather["month_day_max"]=core_weather["month_max"]/core_weather["temp_max"]
core_weather["max_min"]=core_weather["temp_max"]/core_weather["temp_min"]
```

```python
core_weather
```

```python
predictors=["precip","temp_max","temp_min","month_max","month_day_max","max_min"]
error,combined=weather_predictions(predictors,core_weather,model_1)
```
