```python
teams=teams.dropna()
```

```python
teams=teams[["team","country","year","athletes","prev_medals","medals"]]
```

```python
teams
```

```python
teams.year.sort_values().unique()
```

```python
train=teams[teams["year"] <2012].copy()
test=teams[teams["year"] >=2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
np.round(((train.shape[0]/(train.shape[0]+test.shape[0]))*100))
```

```python
reg_model=LinearRegression()
```

```python
predictors=["athletes","prev_medals"]
target="medals"
```

```python
reg_model.fit(train[predictors],train[target])
```

```python
predictions=reg_model.predict(test[predictors])
```

```python
reg_model.score(test[predictors],test[target])
```

```python
test["predictions"]=reg_model.predict(test[predictors])
```

```python
test
```

```python
test.loc[test["predictions"]<=0,"predictions"]=0
```

```python
test["predictions"]=test["predictions"].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(test["medals"],test["predictions"])
error
```

```python
teams.describe()["medals"]
```

```python
test["Match_ind"]="N"
test.loc[test["medals"]==test["predictions"],"Match_ind"]="Y"
```

```python
test.Match_ind.value_counts()
```

```python
errors=(test["medals"]-test["predictions"]).abs()
```

```python
errors
```

```python
error_by_team=errors.groupby(test["team"]).mean()
error_by_team
```

```python
medals=test["medals"].groupby(test["team"]).mean()
medals
```

```python
error_ratio=error_by_team/medals
error_ratio
```

```python
error_ratio=error_ratio[~pd.isnull(error_ratio)]
error_ratio=error_ratio[np.isfinite(error_ratio)]
```
